```python
from datetime import datetime
import cv2
import os

frame_count = 0
save_interval = 30  
save_dir = "smiles"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)


face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml')


cap = cv2.VideoCapture('IMG_1772.MOV')

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to get frame.")
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        pass
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), thickness=2 )
        face_gray = gray[y:y+h, x:x+w]
        face_colour = frame[y:y+h, x:x+h]
        smiles = smile_cascade.detectMultiScale(face_gray, scaleFactor=1.7, minNeighbors=35, minSize=(50, 50))
        is_smiling = len(smiles) > 0

        for (sx, sy, sw, sh) in smiles:
            cv2.rectangle(frame, (x+sx, y+sy), (x+sx+sw, y+sy+sh), (0, 0, 255), 2)
        lbl = "Smile :)" if is_smiling else "No Smile"
        cv2.putText(frame, lbl, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        
        # if is_smiling and frame_count % save_interval == 0:
        #     timestamp = datetime.now().strftime("%Y%m%d_%H%M%S_%f")
        #     filename = f"smiles/smile_{timestamp}.jpg"
        #     cv2.imwrite(filename, face_colour)

    cv2.imshow('Face and Smile Detection', frame)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()



```
